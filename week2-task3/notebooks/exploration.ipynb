# PlaceMux Data Exploration

This notebook demonstrates loading the data and running basic exploration.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

students = pd.read_csv('../data/students.csv')
jobs = pd.read_csv('../data/jobs.csv')
apps = pd.read_csv('../data/applications.csv')
features = pd.read_csv('../data/features.csv')
baseline = pd.read_csv('../data/baseline_results.csv')
```

```python
print(f"Total Students: {len(students)}")
print(f"Total Jobs: {len(jobs)}")
print(f"Total Applications: {len(apps)}")
```

### Baseline Distribution

```python
sns.histplot(data=baseline, x='baseline_score', hue='successful_match', bins=20)
plt.title('Baseline Score Distribution by Success')
plt.show()
```

### Feature Distributions

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.histplot(data=features, x='skill_overlap', hue='successful_match', bins=10, ax=axes[0])
axes[0].set_title('Skill Overlap Distribution')

sns.countplot(data=features, x='experience_match', hue='successful_match', ax=axes[1])
axes[1].set_title('Experience Match vs Success')

sns.countplot(data=features, x='education_match', hue='successful_match', ax=axes[2])
axes[2].set_title('Education Match vs Success')

plt.tight_layout()
plt.show()
```
